```julia
using Plots;
using JLD;
using Printf;
```

```julia
results_path = pwd() * "/plots/";
results_path = results_path * "slow/"

iterations = load(results_path * "iterations.jld")["data"];

avg_compl_time_slow = load(results_path * "avg_comp_time.jld")["data"];
avg_velocity_slow = load(results_path * "avg_velocity.jld")["data"];
rewards_slow = load(results_path * "reward.jld")["data"];
success_rate_slow = load(results_path * "success_rate.jld")["data"];


results_path = pwd() * "/plots/";
results_path = results_path * "fast/"

avg_compl_time_fast = load(results_path * "avg_comp_time.jld")["data"];
avg_velocity_fast = load(results_path * "avg_velocity.jld")["data"];
rewards_fast = load(results_path * "reward.jld")["data"];
success_rate_fast = load(results_path * "success_rate.jld")["data"];
```

```julia
fig_path = pwd() * "/plots/combi/";
```

```julia
ticks = collect(0:10_000_000:40_000_000);
ticklabels = [ @sprintf("%1.2fx10‚Å∂",iterations/1000000) for iterations in ticks ];
```

```julia
plot(iterations, rewards_slow, 
    xlabel="Iterations", 
    ylabel="Reward", 
    label = "slow mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "reward_slow.svg")
```

```julia
plot(iterations, rewards_fast, 
    xlabel="Iterations", 
    ylabel="Reward", 
    label = "fast mode", 
    color = :orange, 
    legend = true, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "reward_fast.svg")
```

```julia
plot(iterations, success_rate_slow, 
    xlabel="Iterations", 
    ylabel="Success Rate", 
    label = "slow mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(iterations, success_rate_fast, 
    xlabel="Iterations", 
    ylabel="Success Rate", 
    label = "fast mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
plot!(yticks=0:0.1:1)
```

```julia
savefig(fig_path * "success_rate.svg")
```

```julia
plot(iterations, avg_velocity_slow, 
    xlabel="Iterations", 
    ylabel="Average Velocity", 
    label = "slow mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(iterations, avg_velocity_fast, 
    xlabel="Iterations", 
    ylabel="Average Velocity", 
    label = "fast mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "avg_velocity.svg")
```

```julia
plot(iterations, avg_compl_time_slow, 
    xlabel="Iterations", 
    ylabel="Average Completion Time", 
    label = "slow mode", 
    legend = true, 
    xformatter = :scientific, 
)
plot!(iterations, avg_compl_time_fast, 
    xlabel="Iterations", 
    ylabel="Average Completion Time", 
    label = "fast mode", 
    legend = true, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "avg_comp_time.svg")
```
